```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize']=8,16
```

```python
#downloading easyocer model and setting it to english
reader = easyocr.Reader(['te'])
```

```python
Image('image.png')
```

```python
#readtext is function in easyocr 
output=reader.readtext('image.png')
# first part bounding box coordinates second part text third part is confidence level
```

```python
output
```

```python
cord = output[0][0]
print(cord)
```

```python
card = output[1][1]
print(card)
```

```python
#bounding boxes
```

```python
x_min, y_min=[int(min(idx)) for idx in zip(*cord)]
x_max, y_max=[int(max(idx)) for idx in zip(*cord)]
```

```python
image=cv2.imread('image.png')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```
